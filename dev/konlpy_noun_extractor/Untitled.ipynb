```python
import pandas as pd
import re

from konlpy.tag import Komoran

from collections import Counter
from itertools import chain
```

```python
nlp = Komoran()
```

```python
raw_df = pd.read_csv("sample_newspaper.csv", sep=',', encoding='utf-8')
raw_list = raw_df['기사']
```

```python
pre_sents = []

for raw_sent in raw_list:
    pre_sent = ' '.join(re.compile('[가-힣0-9]+').findall(raw_sent))
    if len(pre_sent)==0:
        continue
    pre_sents.append(pre_sent)
```
